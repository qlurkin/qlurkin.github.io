#### 2 - Correct the code below (aiming to apply the FFT to the sound signal) and explain the correction done.

Here's the code to correct :

```python
import numpy as np
from matplotlib import pyplot as plt

X = np.fft.fft(noise)
N = len(X)
Amp = abs(X) / N
Amp = Amp[:N//2] * 2
dF = Fs / N
n = np.arange(1, len(Amp))
f = n * dF
plt.stem(f,Amp)
plt.show()
```
