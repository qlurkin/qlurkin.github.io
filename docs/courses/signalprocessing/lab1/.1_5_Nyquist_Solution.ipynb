# Nyquist diagram plotting

Let's suppose the next system : $$ H(s) = \frac{2}{s^2 + 5s + 13} $$

You're asked to :
1. Compute the system frequency response (gain and phase in steady-state as answer to a sine wave),
supposing the input signal's frequency changes from 0.001 to 1000000 rad/s, with 100 logarithmically
spaced data.
1. Plot the Nyquist diagram of this system.
1. Superimpose the computed values (at question 1) on the Nyquist plot. 

```python
import control as ct
import numpy as np
from matplotlib import pyplot as plt

%config InlineBackend.figure_formats = ['svg']

s=ct.tf('s')
assert s is not None

H=2/(s**2+5*s+13)

# Nyquist plotting
plt.figure()
ct.nyquist(H, label='Using control package', title="Nyquist Plot")


# Computing the frequency response by replacing the TF by a complex number
w=np.logspace(-3,6,100)
Hw=2/((1j*w)**2+5*(1j*w)+13)
gain=np.abs(Hw)
ph=np.angle(Hw)

# Superimposition of the curves
plt.plot(gain*np.cos(ph), gain*np.sin(ph), '*r', label='With calculation')
plt.legend()
plt.show()
```
